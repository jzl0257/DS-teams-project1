# 22-eda-soil

In this notebook, we will do exploratory data analysis on the cleaned data.

```python
import pandas as pd
soil=pd.read_csv("ArchaeologicalSoilData.csv")
soil.head()
```

```python
soil.info()
```

```python
soil=soil.iloc[1:,].reset_index(drop=True)
soil
```

```python
soil[["Area","Perimeter","Volume","Da","Dp"]]=soil[["Area","Perimeter","Volume","Da","Dp"]].astype("float")
soil.info()
```

# Correlation Plot and Matrix

```python
corr=soil[["Area","Perimeter","Volume","Da","Dp"]].corr()
corr
```

```python
import seaborn as sns
sns.heatmap(corr)
```

# Boxplot

```python
from matplotlib import pyplot as plt
_, axss=plt.subplots(2,3,figsize=[20,10])
sns.boxplot(soil["Area"],ax=axss[0][0])
sns.boxplot(soil["Perimeter"],ax=axss[0][1])
sns.boxplot(soil["Volume"],ax=axss[0][2])
sns.boxplot(soil["Da"],ax=axss[1][0])
sns.boxplot(soil["Dp"],ax=axss[1][1])
```
