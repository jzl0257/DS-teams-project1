In this notebook, we will do exploratory data analysis on the cleaned data.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

lmd = pd.read_csv("LithicExperimentalData.csv")
lmd.head()
```

```python
lmd.info()
```

```python
# area vs perimeter
sns.lmplot(x = 'Area',y = 'Perimeter', hue='label', data = lmd)
plt.title('Relationship between Area and Perimeter')
```

```python
# Dp
sns.stripplot(x = 'label',y = 'Dp', hue= "label", data = lmd, jitter=True)
sns.boxenplot(x = "label", y = "Dp",data = lmd)
plt.xlabel('Lithic')
plt.title('Dp: perimeter equivalent diameter')
```

```python
# Da
sns.stripplot(x = 'label',y = 'Da', hue= "label", data = lmd, jitter=True)
sns.boxenplot(x = "label", y = "Da",data = lmd)
plt.xlabel('Lithic')
plt.title('Da: area equivalent diameter')
```

```python
# Da vs Dp
sns.lmplot(x = 'Da',y = 'Dp', hue='label', data = lmd)
plt.title('Relationship between Da and Dp')
```

```python
plt.scatter(lmd["FWidth"],lmd["EWidth"],cmap = "coolwarm")
plt.xlabel('FWidth')
plt.ylabel('EWidth')
plt.title('FWidth vs EWidth')
```

```python
plt.scatter(lmd["FLength"],lmd["ELength"],cmap = "coolwarm")
plt.xlabel('FLength')
plt.ylabel('ELength')
plt.title('FLength vs ELength')
```

```python
plt.scatter(lmd["FThickness"],lmd["EThickness"],cmap = "coolwarm")
plt.xlabel('FThickness')
plt.ylabel('EThickness')
plt.title('FThickness vs EThickness')
```

```python
plt.scatter(lmd["Concavity"],lmd["Convexity"],cmap = "coolwarm")
plt.xlabel('Concavity')
plt.ylabel('Convexity')
plt.title('Concavity vs Convexity')
```

```python
plt.scatter(lmd["CHull  Area"],lmd["CHull Perimeter"],cmap = "coolwarm")
plt.xlabel('CHull  Area')
plt.ylabel('CHull Perimeter')
plt.title('CHull  Area vs CHull Perimeter')
```

```python
plt.scatter(lmd["Circularity"],lmd["Roundness"],cmap = "coolwarm")
plt.xlabel('Circularity')
plt.ylabel('Roundness')
plt.title('Circularity vs Roundness')
```
